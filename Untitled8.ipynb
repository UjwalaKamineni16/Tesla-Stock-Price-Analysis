```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from arch import arch_model
from textblob import TextBlob
import yfinance as yf
from sklearn.metrics import mean_squared_error
```

The Tesla stock price dataset was obtained from the WRDS Account, ensuring high-quality and reliable data. Yahoo Finance data was used to extract recent news headlines for sentiment analysis, as it provides accessible and relevant information.


```python
# Loading the Tesla stock prices dataset
file_path = 'TeslaStockPricesDaily.csv'  
tesla_data = pd.read_csv(file_path)
tesla_data.head()
```

```python
# Data Preprocessing
tesla_data['date'] = pd.to_datetime(tesla_data['date'])
tesla_data.sort_values(by='date', inplace=True)
tesla_data.rename(columns={'PRC': 'Close', 'VOL': 'Volume', 'RET': 'Daily_Return'}, inplace=True)
```

```python
tesla_data.columns
```

```python
# Handlimg Missing Values
tesla_data['Daily_Return'] = tesla_data['Daily_Return'].fillna(0)

# Calculations
tesla_data['MA_20'] = tesla_data['Close'].rolling(window=20).mean()
tesla_data['MA_50'] = tesla_data['Close'].rolling(window=50).mean()
tesla_data['MA_200'] = tesla_data['Close'].rolling(window=200).mean()
tesla_data['Volatility'] = tesla_data['Daily_Return'].rolling(window=20).std()
tesla_data['Cumulative_Return'] = (1 + tesla_data['Daily_Return']).cumprod()
```

```python
# Correlation Analysis
correlation_matrix = tesla_data[['Close', 'Volume', 'Daily_Return', 'Volatility']].corr()

# Print Correlation Matrix
print("Correlation Matrix:")
print(correlation_matrix)
#Heatmap for Visualization
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Tesla Stock Data")
plt.show()
```

```python
# Plotting
fig, axes = plt.subplots(5, 1, figsize=(12, 25))

# 1. Tesla Stock Price with Moving Averages
axes[0].plot(tesla_data['date'], tesla_data['Close'], label='Close Price')
axes[0].plot(tesla_data['date'], tesla_data['MA_20'], label='20-Day MA')
axes[0].plot(tesla_data['date'], tesla_data['MA_50'], label='50-Day MA')
axes[0].plot(tesla_data['date'], tesla_data['MA_200'], label='200-Day MA')
axes[0].set_title('Tesla Stock Price with Moving Averages')
axes[0].legend()
axes[0].grid()

# 2. Tesla Trading Volume
axes[1].bar(tesla_data['date'], tesla_data['Volume'], color='gray')
axes[1].set_title('Tesla Trading Volume')
axes[1].grid()

# 3. Volatility
axes[2].plot(tesla_data['date'], tesla_data['Volatility'], color='red')
axes[2].set_title('Tesla Stock Volatility (20-Day Rolling)')
axes[2].grid()

# 4. Cumulative Returns
axes[3].plot(tesla_data['date'], tesla_data['Cumulative_Return'], color='green')
axes[3].set_title('Tesla Cumulative Returns')
axes[3].grid()

# 5. Daily Returns Histogram
axes[4].hist(tesla_data['Daily_Return'].dropna(), bins=50, color='blue', alpha=0.7)
axes[4].set_title('Distribution of Tesla Daily Returns')
axes[4].grid()

# Final Show
plt.tight_layout()
plt.show()
```

# The ARIMA model is used to forecast Tesla's stock price to test the hypothesis that Tesla's price follows predictable patterns over time.


```python
# Time Series Analysis using ARIMA
train_size = int(len(tesla_data) * 0.9)
train, test = tesla_data['Close'][:train_size], tesla_data['Close'][train_size:]

# Fit ARIMA Model
arima_model = ARIMA(train, order=(1, 1, 1))
fit_arima = arima_model.fit()

# Forecast and Plot Results
forecast = fit_arima.forecast(steps=len(test))
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"Test RMSE: {rmse:.2f}")

plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Actual Test Data')
plt.plot(test.index, forecast, label='ARIMA Forecast', linestyle='--')
plt.legend()
plt.title('ARIMA Model - Tesla Stock Price Forecast')
plt.show()
```

*"The ARIMA model forecast aligns with the test data, validating the hypothesis of predictable stock price trends with an RMSE of 49.47."*


# The GARCH model analyzes Tesla's volatility, testing whether increased volatility influences cumulative returns.


```python
# GARCH Model for Volatility Forecasting
returns = tesla_data['Daily_Return'].dropna()
garch_model = arch_model(returns, vol='Garch', p=1, q=1, mean='Zero')
garch_fit = garch_model.fit(disp='off')
print(garch_fit.summary())

# Forecast Volatility for 30 Days
forecast_horizon = 30
garch_forecast = garch_fit.forecast(horizon=forecast_horizon)
future_volatility = np.sqrt(garch_forecast.variance.values[-1])

plt.figure(figsize=(12, 6))
plt.plot(returns.index[-100:], returns[-100:], label='Recent Daily Returns')
plt.plot(range(len(returns), len(returns) + forecast_horizon), future_volatility, label='Forecast Volatility', color='red')
plt.title('GARCH Model - Volatility Forecast')
plt.legend()
plt.show()
```

```python
# Sentiment Analysis with Yahoo Finance
tesla_ticker = yf.Ticker("TSLA")
tesla_news = tesla_ticker.news

# Extract News Titles and Dates
news_titles = [news['title'] for news in tesla_news]
news_dates = [pd.to_datetime(news['providerPublishTime'], unit='s') for news in tesla_news]

# Perform Sentiment Analysis
news_df = pd.DataFrame({'Date': news_dates, 'Title': news_titles})
news_df['Sentiment'] = news_df['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)
news_df['Sentiment_Label'] = news_df['Sentiment'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

# Display Sentiment Analysis Results
print("Tesla News Sentiment Analysis:\n", news_df)

# Sentiment Visualization
plt.figure(figsize=(8, 6))
news_df['Sentiment_Label'].value_counts().plot(kind='bar', color=['green', 'red', 'gray'])
plt.title("Tesla News Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# Sentiment Scores Over Time
plt.figure(figsize=(12, 6))
plt.plot(news_df['Date'], news_df['Sentiment'], marker='o', linestyle='', color='orange')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Tesla News Sentiment Over Time")
plt.xlabel("Date")
plt.ylabel("Sentiment Score")
plt.show()
```

*"The sentiment analysis shows a dominance of positive news sentiment, which may reflect bullish behavior in Tesla's stock price."*
